```python
import pandas as pd
import numpy as np
import re
import openai
from config import *
```

```python
branches_df = pd.read_csv('../../raw_data/branches_raw.csv')
reviews_df = pd.read_csv('../../raw_data/reviews_raw.csv')
```

```python
openai.api_key = api_key
```

```python
def get_analysis(review):
    prompt = """
        This text is a bunch of reviews for a bank branch. 
        Analyze the text without mentioning names of people. 
        I repeat. Do not mention personal names of employees. 
        Bullet point the positives, room for improvement, and miscellaneous: 
        """
    prompt += str(review)
    
    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo-16k-0613',
        messages=[
            {'role': 'user', 'content': prompt}
        ]
    )
    
    return response['choices'][0]['message']['content']    
```

# Branches

```python
# filters reviews to those that occurred within the past 1-2 years
past_year = reviews_df['date'].str.contains('year ago|month|week|day').fillna(False)
branches_df['review'] = reviews_df[past_year].groupby('branch_id', as_index=False)['review'].sum()['review']
branches_df['analysis'] = ''

# reviews_df['ts_review'] = reviews_df['date'] + ': ' + reviews_df['review']
# branches_df['review'] = reviews_df.groupby('branch_id', as_index=False)['ts_review'].sum()['ts_review']
```

```python
i = 0
```

```python
for index, row in branches_df.iloc[i:].iterrows():
    print(i)
    if pd.notnull(row['review']):
        branches_df.at[index, 'analysis'] = get_analysis(row['review'])
    i+= 1
```
